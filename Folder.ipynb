```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    function = (-12 * x ** 4) * np.sin(np.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
    return function
```

```python
x_limit = [-10, 10]
x = np.arange(x_limit[0], x_limit[1], 0.1)
color = 'r'
x_change = []
x_root = []
func_direct = -1
```

```python
for i in range(len(x) - 1):
    if func_direct == -1:
        if func(x[i]) < func(x[i] + 1):
            func_direct = 1
            x_change.append((x[i], func(x[i])))
    else:
        if func(x[i]) > func(x[i] + 1):
            func_direct = -1
            x_change.append((x[i], func(x[i])))

for i in range(len(x) - 1):
        if (func(x[i]) > 0 and func(x[i] + 1) < 0) or (func(x[i]) < 0 and func(x[i] + 1) > 0):
            x_root.append((x[i], func(x[i])))
```

```python
print(len(x_root), x_root)
```

```python
print(len(x_change), x_change)
```

```python
def change_color():
    global color
    if color == 'r':
        color = 'b'
    else:
        color = 'r'
    return color
```

```python
x_cur = np.arange(x_limit[0], x_change[0][0], 0.1)
plt.plot(x_cur, func((x_cur)), change_color())
for i in range(len(x_change) - 1):
    x_cur = np.arange(x_change[i][0], x_change[i + 1][0], 0.1)
    plt.plot(x_cur, func((x_cur)), change_color())
x_cur = np.arange(x_change[-1][0], x_limit[1], 0.1)
plt.plot(x_cur, func((x_cur)), change_color())

for x_cur in x_root:
    plt.plot(x_cur[0], func(x_cur[0]), 'go')
plt.grid(visible=True)


plt.show()
```
